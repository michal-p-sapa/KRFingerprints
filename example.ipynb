```python
import pandas as pd
import numpy as np
from KRFingerprints import *
```

```python
dataset = pd.read_csv('dataset.csv')
dataset
```

```python
dataset_fps = KRFingerprints.MultipleDescriptorsDataFrame(dataset,'Smiles')
```

```python
dataset_fps
```

```python
KRFingerprints.DrawFingerprints(['KRFP4645', 'KRFP3321', 'KRFP2137'])
```

```python
selected = dataset_fps[dataset_fps['KRFP4645']==1]
selected
```

```python
import rdkit.Chem as Chem
KRFingerprints.HighlightKRFP([Chem.MolFromSmiles(x) for x in selected['Smiles']], krfp='KRFP4645', names=list(selected['ZINC_ID']))
```
